```python
import numpy as np
import matplotlib.pyplot as plt

test_ensemble = np.random.normal(0., 1., 5000)

test_ensemble_average = np.sum(test_ensemble)/5000.
test_ensemble_stdev = np.std(test_ensemble)

print test_ensemble_average
print test_ensemble_stdev

plt.hist(test_ensemble)
plt.show()

test_mult = test_ensemble * test_ensemble

test_mult_average = np.sum(test_mult)/5000.
test_mult_stdev = np.std(test_mult)

print test_mult_average
print test_mult_stdev

plt.hist(test_mult)
plt.show()
```
